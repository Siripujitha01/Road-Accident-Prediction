```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_excel('./Project Dataset.xlsx')
df.head()
```

```python
x  = pd.get_dummies(df.iloc[:,1:-1])
y  = df[['AccidentOccurance ']]
```

```python
x.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
X_train.head()
```

```python
y_train.tail()
```

```python
X_test.head()
```

```python
y_test['AccidentOccurance '].head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
predict = lr.predict(X_test)
predict
```

```python
p=1/(1+np.e**(0-X_test['Vehicle_Age']))
```

```python
plt.scatter(X_test['Vehicle_Age'], p, color='red')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
matrix =pd.DataFrame(confusion_matrix(y_test, predict), columns = ['Predicted No', 'Predicted Yes'], index = ['Actual No', 'Actual Yes'])
matrix
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test, predict))
print("Accuracy Score :",accuracy_score(y_test, predict))
```
